<h3> Average Price by Neighbourhood Group

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded the Airbnb data into the airbnb_data DataFrame

# Group by 'neighbourhood_group' and calculate the average price for each group
average_prices = airbnb_data.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)

# Create a horizontal bar plot using seaborn with a red-orange-pink palette
plt.figure(figsize=(10, 6))
sns.barplot(x=average_prices.values, y=average_prices.index, palette='Reds_r')  # '_r' reverses the palette
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood Group')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x="neighbourhood_group", y="price", data=airbnb_data, estimator=lambda x: sum(x)/len(x))
plt.title("Average Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")
plt.show()
```

6. **50th Percentile (50% or Median):**

7. **75th Percentile (75% or Third Quartile):**

These summary statistics provide a concise overview of the central tendency, spread, and distribution of prices in the Airbnb dataset. They help to understand the typical price range, the presence of outliers, and the overall variability in the data.

<h3> Sentiment Analysis of Name

```python
import pandas as pd
import seaborn as sns

# Calculate the correlation coefficient between number_of_reviews and price
corr = airbnb['number_of_reviews'].corr(airbnb['price'])
print(corr)

# Create a correlation matrix
corr_matrix = pd.DataFrame({'number_of_reviews': airbnb['number_of_reviews'], 'price': airbnb['price']}).corr()

# Create a heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Set the title
plt.title(f'Correlation between Number of Reviews and Price (corr={corr:.2f})')

# Display the plot
plt.show()
```

there is a weak negative relationship between the two variables

A 1/x graph is the inverse of a linear relationship, meaning that as one variable increases, the other decreases exponentially. This could indicate that there are diminishing returns to the number of reviews, where additional reviews have less impact on price as the number of reviews increases

I think the threshold where additional reviews have less impact on price is around 200
