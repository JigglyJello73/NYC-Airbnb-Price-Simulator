<h3> sentiment 

```python
from textblob import TextBlob

# Function to get sentiment polarity from text
def get_sentiment(text):
    analysis = TextBlob(str(text))
    return analysis.sentiment.polarity

# Apply the sentiment analysis function to the 'name' column and create a new 'sentiment' column
df['sentiment'] = df['name'].apply(get_sentiment)

# Display the updated DataFrame with the new 'sentiment' column
df.head(3)
```

<h3> find any latitude and longitude in df inside central park and remove them from df



```python
# Assuming df is your DataFrame
central_park_latitude = 40.7826
central_park_longitude = -73.9656

# Define a radius (you may adjust this based on the desired range)
radius = 0.01  # Example: 0.01 degree

# Filter rows outside the specified latitude and longitude range
df_filtered = df[
    ~(
        (df['latitude'] >= central_park_latitude - radius) &
        (df['latitude'] <= central_park_latitude + radius) &
        (df['longitude'] >= central_park_longitude - radius) &
        (df['longitude'] <= central_park_longitude + radius)
    )
]

# Reset the index after filtering
df_filtered = df_filtered.reset_index(drop=True)
df = df_filtered

# Display the updated DataFrame
df.shape
```
